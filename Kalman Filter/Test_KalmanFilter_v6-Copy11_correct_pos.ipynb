```python
# Initializze state matrices
X = np.array([[100], [0]])
# P0 = np.diag((0.01, 0.01, 0.01, 0.01))
P = np.diag((0.01,0.01))
# A = np.array([[1, 0, Ts , 0], [0, 1, 0, Ts], [0, 0, 1, 0], [0, 0, 0, 1]])
Q = np.eye(X.shape[0])
# Q = np.zeros((2,2))
# B = np.eye(X.shape[0])
# U = np.zeros((X.shape[0],1)) 

# Measurement matrices
Y = np.array([[X[0,0] + abs(np.random.randn(1)[0])], [X[1,0] + abs(np.random.randn(1)[0])]])
H = np.array([[1, 0], [0, 1]])
R = np.eye(Y.shape[0]) * 1
# R = np.array([[1,0],[0,0]])

```

```python
# # System matrix - state
A = np.array([[1, Ts], [0,1]])

# # System matrix - input
B= np.array([[-0.5*(Ts**2)], [-Ts]])
# # G = [[-0.5*(Ts**2)], [-Ts]]

# # Input vector
# u = 9.80665
# U = np.array([[0],[9.80665]])
U = 9.80665

# # Observation matrix
# H = np.array([1,0])

# # Sigma - Standard Deviations
# Q = np.array([[0,0],[0,0]])
```

```python
# Apply Kalman Filter
xx = []

measure_x_0 = []
measure_y_0 = []

for i in range(0,N):
    
    # Prediction
    X = np.dot(A,X) + np.dot(B.T,U)
    P = np.dot(A,np.dot(P, A.T)) + Q
    xx.append(X)
    plt.plot((X[0,0]), (X[1,0]), 'o', color='red')
    
    # Update
    nu = np.dot(H,X)
    S = np.dot(H, np.dot(P,H.T)) + R
    K = np.dot(P, np.dot(H.T, np.linalg.pinv(S)))
    X = X + np.dot(K, Y-nu)
    P = P - np.dot(K, np.dot(S,K.T))
    plt.plot(X[0,0], X[1,0], 'o', color='black')
    
    Y = np.array([[X[0,0] + abs(0.1 * np.random.randn(1)[0])],[X[1, 0] + abs(0.1 * np.random.randn(1)[0])]]) 
    measure_x_0.append(Y[0,0])
    measure_y_0.append(Y[1,0]) 
    plt.plot(Y[0,0], Y[1,0],'+', color='green')
    
# Y = np.array([[X[0,0] + abs(0.1 * np.random.randn(1)[0])],[X[1, 0] + abs(0.1 * np.random.randn(1)[0])]]) 
# plt.plot(Y[0,:], Y[1,:], color='blue')

# plt.plot(sample_range2,measure_x_0)
```

```python
M = 10
N = 100
R1 = 0
R2 = 1
nR = np.linspace(R1, R2, M)
nT = np.linspace(0, 2*np.pi,N)
R,T = np.meshgrid(nR, nT)
```

```python
#%%
# Convert grid to cartesian coordinates
X = R * np.cos(T)
Y = R * np.sin(T)
Z = np.ones((len(T),1)) 
row, col = X.shape

```
